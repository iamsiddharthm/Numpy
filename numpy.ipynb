###Array reshaping 

```python
import numpy as np

arr = np.arange(12)
print("original output:", arr)

reshaped = arr.reshape((3, 4))  
print("reshaped output:\n", reshaped)

flattend=reshaped.flatten()
print("\nFlattened array is ",flattend)

ravel=reshaped.ravel()
print("\n raveled :",ravel)

# ravel return view(original array ) instead of copy
```
